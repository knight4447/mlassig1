# 1- Using Make_blob generate data of 1000 data points with three cluster apply kmeans on it with k = 3 and use the metrics and get the accuracy (For Accuracy take reference of DBSCAN evaluation)
● Apply DBscan on Cust Segmentation Data

```python
from sklearn.datasets import make_blobs
```

```python
# Blob Genearation
x, y = make_blobs(n_samples=1000, centers=3, n_features=2,random_state=100)
x
```

```python
y
```

```python
from sklearn.cluster import KMeans


kmeans = KMeans(n_clusters=3, random_state=100)
y_pred = kmeans.fit_predict(x)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(x, y_pred, metric='euclidean')
```

```python
import pandas as pd 
```

```python
df = pd.read_csv('Copy of Cust_Segmentation.csv')
df.head()
```

```python
df.columns
```

```python
x_n = df[['Age', 'Edu', 'Years Employed', 'Income', 'Card Debt',
       'Other Debt', 'DebtIncomeRatio']]
y_n = df['Defaulted']
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=13, min_samples=5)
y_pred = dbscan.fit_predict(x_n)
y_pred
```

```python
silhouette_score(x_n, y_pred, metric='euclidean')
```

# 2-Using dirtydata.csv Demonstrate all the techniques for removing the null values

● Replace by MEAN

● Replace by MEDIAN

● Replace by MODE

● Replace by ARBITUARY VALUE

● Replace by 0

```python
dd = pd.read_csv('dirtydata (1).csv')
dd
```

```python
dd.isnull().sum()
```

```python
mean = dd['Calories'].mean()
dd['Calories'].fillna(mean, inplace = True)

dd
```

```python
# Replace null values by median
median = dd['Calories'].median()
dd['Calories'].fillna(median, inplace=True)

dd
```

```python
# Replace null values by mode
mode = dd['Date'].mode()[0]
dd['Date'].fillna(mode, inplace=True)

dd
```

```python
# Replace null values by arbitrary value (100)
cal = dd['Calories'].fillna(100, inplace=True)

dd
```

```python
# Replace null values by 0
null = dd.fillna(0, inplace=True)
dd
```
