```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import VarianceThreshold
```

```python
corr_matrix = df.corr()
plt.subplots(figsize=(50,45))
sns.heatmap(corr_matrix,annot=True)
```

```python
var_threshold = 0.1
selector = VarianceThreshold(var_threshold)
selector.fit_transform(df)
```

```python
important_cols = df.columns[selector.get_support()]
important_cols = corr_matrix[abs(corr_matrix) > 0.4].stack().reset_index().\
                query('level_0 != level_1').level_0.unique()
df = df[important_cols]
```

```python
df
```

```python
set(df['attack'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = df.drop('attack', axis=1)
y = df['attack']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state = 42)
```

```python


from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler    
#st_x= StandardScaler()    
#x_train= st_x.fit_transform(X_train)    
#x_test= st_x.transform(X_test)  

lr =LogisticRegression(solver='liblinear')
lr.fit(X_train, y_train)
ypred = lr.predict(X_test)
accuracy = accuracy_score(y_test, ypred)
```

```python
print("Logistic Regression: ")
print("Accuracy -" ,accuracy*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
ypred1 = rf.predict(X_test)
acc = accuracy_score(y_test, ypred1)
print("Random Forest Classifier")
print("Accuracy -" ,acc*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
ypred2 = dt.predict(X_test)
accu = accuracy_score(y_test, ypred2)
print("Decision Tree ")
print("Accuracy -" ,accu*100)
```
